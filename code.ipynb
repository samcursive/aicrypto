# Convert CipherBank into one dataset. 

```python
# Change column names to prevent confusion
# atbash_df = atbash_df.rename(columns={"":"ciphertext"})
# dualavg_df 
# parityshift_df 

atbash_df = atbash_df.rename(columns={"Atbash":"ciphertext"})
atbash_df["algorithm"] = "AtBash"
atbash_df.head(3)

```

```python
cols = ["domain", "sub_domain", "plaintext", "ciphertext", "algorithm"]



# atbash_df = []
# dualavg_df = dualavg_df[columns]
# parityshift_df 
atbash_df.head(3)

# master_df = pd.concat([])
```

```python
dualavg_df = dualavg_df.rename(columns={"DualAvgCode":"ciphertext"})
dualavg_df["algorithm"] = "Dual Average"
dualavg_df.head(3)
```

```python
parityshift_df = parityshift_df.rename(columns={"ParityShift":"ciphertext"})
parityshift_df["algorithm"] = "Parity Shift"
parityshift_df.head(3)
```

```python
cols
atbash_df = atbash_df[cols]
dualavg_df = dualavg_df[cols]
parityshift_df = parityshift_df[cols]

master_df = pd.concat([atbash_df, dualavg_df, parityshift_df ], ignore_index=True)
# print(f"Rows per algorithm:\n,{master_df["algorithm"].value_counts()}")
print(f"Rows per algorithm:\n{master_df['algorithm'].value_counts()}")
print(f"\nAny missing ciphertext? {master_df['ciphertext'].isna().any()}")
master_df.head(780)


#Save Master Dataset
try:
    master_df.to_csv("datasets/CipherBank_Master.csv", index=False) 
    print("Dataset Saved successfully.")
    save_status=1
except Exception as e:
    print(f"Error saving Dataset: '{e}.'")
    save_status=0
```
