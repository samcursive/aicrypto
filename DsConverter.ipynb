```python
#Cipher.Atbash Dataset
#ds = load_dataset("yu0226/CipherBank")
bash_ds = load_dataset("yu0226/CipherBank", "Atbash")#, "Atbash")
print(f"Atbash Subset:\n{bash_ds}")
print(f"Atbash Subset:\n{bash_ds.keys()}\n")


#Cipherbank.Atbash Dataset
dual_average_ds = load_dataset("yu0226/CipherBank", "DualAvgCode")
print(f"Dual Average:\n{dual_average_ds}")
print(f"Dual Average:\n{dual_average_ds.keys()}\n")


#Cipherbank.ParityShift Dataset
parity_shit_ds = load_dataset("yu0226/CipherBank", "ParityShift")
print(f"ParityShift:\n{parity_shit_ds}")
print(f"ParityShift:\n{parity_shit_ds.keys()}\n")
```
